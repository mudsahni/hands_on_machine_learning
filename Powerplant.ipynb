```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
powerplant = pd.read_excel('Folds5x2_pp.xlsx')
powerplant.columns = ["Temperature", "Exhaust Vacuum", "Ambient Pressure", "Relative Humidity", "Energy Output"]
powerplant.head()
```

```python
powerplant.info()
```

```python
powerplant.describe()
```

```python
powerplant.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(powerplant, test_size=0.2, random_state=42)
print(len(train_set), len(test_set))
```

```python
powerplant_exploration = powerplant.copy()
```

```python
powerplant_exploration.plot(kind="scatter", x="Energy Output", y="Temperature", alpha=0.1)
plt.show()
```

```python
powerplant_exploration.plot(kind='scatter', x='Energy Output', y='Temperature', alpha=0.4,
                         s=powerplant_exploration['Ambient Pressure']/100, label='Ambient Pressure',
                         figsize=(15,10), c='Energy Output', cmap=plt.get_cmap('jet'),
                         colorbar=True)
plt.legend()
plt.show()
```
