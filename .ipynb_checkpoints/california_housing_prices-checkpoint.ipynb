## Big Picture
### Objective
Predict median housing price given other data about that district.

**Dataset**: California housing district census.

### Performance Metric
**Root Mean Square Error** (also called the Euclidean distance or the $L_2$ norm): 

$$RMSE(X, h) = \sqrt{\frac{1}{m} \sum^m_{i=1} (h(x^{(i)})-y^{(i)})^2}$$

It is basically the average of squared errors between each prediction. We then square root the average in order to bring it back to the units of the data.

$h: hypothesis \  function \ (prediction)$

$y: ground \ truth$

RMSE is the preferred performance measure for simple regression problems, however, there are instances where you can use another measure. For example, if the data has a lot of outliers, then we can use **Mean absolute deviation** (also called the Manhattan distance or the $L_1$ norm). 

The RMSE, because it is squaring the distance between our prediction and ground truth, is more susceptible to outliers. 

$$MAD(X,h) = \frac{1}{m}\ {\big{|}} h(x^{(i)})-y^{(i)} {\big{|}}$$

The higher the norm index, the more it susceptible it is to outliers. But when outliers are rare, like in a normal distribution, the RMSE performs really well.

```python
# Get data function
import os
import tarfile
from six.moves import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml/master/" 
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()

fetch_housing_data()
```

### Creating a test set
We want our machine learning algorithm to be able to generalize. For that it is important that we test the algorithms strength on unseen data. It is important that this data be different from the data we use to train our algorithm on, otherwise it might not generalize and will overfit the training data.

```python
import numpy as np

np.random.seed(42)

def split_train_test(data, test_ratio):
    # get random indices
    shuffled_indices = np.random.permutation(len(data))
    # get test set size
    test_set_size = int(len(data) * test_ratio)
    # test indices
    test_indices = shuffled_indices[:test_set_size]
    # train indices
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]

train_set, test_set = split_train_test(housing, 0.2)
print(len(train_set), len(test_set))
```

However, this approach is not perfect. Since the indices are always random, if we run this function again, we will get a new dataset and this means over time, our machine learning algorithm can get to see the whole dataset.

One solution is to save the test set generated in the first run and then keep using that. Another, is to set the seed so that the same random numbers are generated at every run. But both these solutions will break once we get an updated dataset. 

Another solution is to use an identifier for each instance of data. Using this we can ensure that the new test set will contain 20% of the data but not contain any data which was in the training set.

```python
from zlib import crc32

def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2 ** 32 

def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

The housing dataset does not have an identifier column, so we can use a combination of the latitude and longitude columns.

```python
housing_with_id = housing.copy()
housing_with_id['id'] = housing['longitude'] * 1000 + housing['latitude']
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, 'id')
print(len(train_set), len(test_set))
```

### Sampling

So far we have considered purely random methods, which are fine if we have a large amount of data. However, if we have a small dataset then we run the risk of introducting a significant sampling bias.

In statistical analysis, it is extremely important to have a significantly large random sample so that our data is spread out across the underlying distribution. However, when using small samples, it is necessary to represent all sections of the distribution proportionally for the analysis. For example, if a survey company calls 1000 people for a poll, they would want to make sure those 1000 people represent all sections of society proportionally rather than just have a random sample. This is called _stratified sampling_.

The population is divided into homogeneous groups called strata and the right number of instances is sampled from each stratum to guarantee that the test set is representative of the overall population. 

Median income is an important attribute to predicting the median housing prices. We want to ensure that the test set is representative of the various categories of incomes in the whole dataset. Since median income is a continuous variable, we will have to convert it to a categorical variable based on a range of income. To understand the range of median income, let's create a histogram.

```python
housing['income_cat'] = pd.cut(housing['median_income'], bins=[0., 1.5, 3.0, 4.5, 6., np.inf], labels=[1,2,3,4,5])
housing['income_cat'].hist()
plt.show()
```

We can use these as income categories and create a stratified sample.

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing['income_cat']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

### Custom Transformers

```python
from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, population_ix] / X[:, households_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
    
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attributes = attr_adder.transform(housing.values)
```

### Feature Scaling
With a few exceptions, Machine Learning algorithms dont perform very well when the numerical attributes have very different scales. This is the case with the housing data: the total number of rooms ranges from about 6 to 39,320, while the median incomes only range from 0 to 15. Note that scaling the target values is generally not required.

There are two common ways to get all attributes to have the same scale: min-max scaling (also called _normalization_) and _standardization_.

- **Normalization**: Values are shifted and rescaled so they end up ranging from 0 to 1. This is done by subtracting the minimum and dividing by the difference between the maximum and minimum values. 

$$ Norm(X) = \frac{X - X_{min}}{X_{max} - X_{min}}$$

- **Standardization**: Values are converted to their _z-scores_ from the mean. It is done by subtracting the mean and then dividing by the standard deviation. This makes the resulting distribution have unit variance. Unlike normalization, standardization does not bind values to a specific range. This may be a problem for some algorithms (eg. neural networks often except an input value ranging from 0-1). However, standardization is much less affected by outliers. 

$$ Z(X) = \frac{X-X_{mean}}{X_{std}}$$

### Transformation Pipelines
Scikit-learn's pipeline constructor takes a list of name/estimator pairs defining a sequence of steps. All but the last estimator but must transformers (i.e. they must have a fit transform method). 

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([('imputer', SimpleImputer(strategy="median")),
                         ('attribs_adder', CombinedAttributesAdder()),
                         ('std_scaler', StandardScaler())])

housing_num_tr = num_pipeline.fit_transform(housing_num)
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num)

cat_attribs = ['ocean_proximity']

full_pipeline = ColumnTransformer([('num', num_pipeline, num_attribs),
                                   ('cat', OneHotEncoder(), cat_attribs)])

housing_prepared = full_pipeline.fit_transform(housing)
```

### Model Selection: Linear Regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(housing_prepared, housing_labels)
```

```python
some_data = housing.iloc[:5]
some_labels = housing_labels[:5]
some_data_prepared = full_pipeline.transform(some_data)
print(f"Predictions: {lin_reg.predict(some_data_prepared)}")
```

```python
print(f"Labels: {list(some_labels)}")
```

It seems to work but it doesnt seem very accurate. Let's measure the root mean square error (RMSE) for this set of predictions.

$$ RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^n (y_i - \hat{y_i})^2}$$

```python
from sklearn.metrics import mean_squared_error
housing_predictions = lin_reg.predict(housing_prepared)
lin_mse = mean_squared_error(housing_labels, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
print(f"RMSE: {lin_rmse}")
```

This means that on average, our predictions has been off by \$68911. Let's calculate the $R^2$ value which is a relative measure of how successfull our predictions have been.

$$ R^2 = 1 - \frac{SSE}{TSE}$$

$$ SSE = RMSE^2 \times n = \sum_{i=1}^n (y_i - \hat{y_i})^2$$

$$ TSS = \sum_{i=1}^n (y_i - \bar{y})^2$$

```python
def r2(prediction, labels):
    n = len(prediction)
    sse = np.sum((prediction - labels)**2)
    tss = np.sum((prediction - np.mean(prediction))**2)
    return 1 - (sse/tss)

print(f"R2 score: {r2(housing_predictions, housing_labels)}")
```

This implies that our input variables are able to predict 45% of variance within the data or that the input variables account for 45% of the output.

### Evaluation: Cross-Validation
k-fold cross validation splits our training set into k folds --> picks a fold for k iterations --> trains on the remaining k-1 folds and then evaluates the remaining data.

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, housing_prepared, housing_labels, scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)

def display_scores(scores):
    print(f"Scores: {scores}")
    print(f"Mean: {scores.mean()}")
    print(f"Standard Deviation: {scores.std()}")
```

```python
display_scores(tree_rmse_scores)
```

The performance of the decision tree has dropped drastically! It, in fact, performs worse than the linear regression model. Cross validation allows us to not only get an estimate of our model, but also a measure of how precise the estimate is (i.e. standard deviation). 

Let's perform cross validation for the linear regression model.

```python
scores = cross_val_score(lin_reg, housing_prepared, housing_labels, scoring='neg_mean_squared_error', cv=10)
lin_rmse_scores = np.sqrt(-scores)
```

```python
display_scores(lin_rmse_scores)
```

### Fine tuning our model
Let's assume that we have a shortlist of promising models. We need to fine-tune them.

### Grid Search
One way to do that would be to fiddle with the hyperparameters manually, until we find a great combination of hyperparameter values. This would be very tedious work, and we may not have time to explore many combinations.

Here we can use Grid Search. 

```python
from sklearn.model_selection import GridSearchCV

param_grid = [ { 'n_estimators' : [ 3 , 10 , 30 ], 'max_features' : [ 2 , 4 , 6 , 8 ]}, 
              { 'bootstrap' : [ False ], 'n_estimators' : [ 3 , 10 ], 'max_features' : [ 2 , 3 , 4 ]}, ]
```

```python
forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)
grid_search.fit(housing_prepared, housing_labels)
```

```python
grid_search.best_params_
```

```python
param_grid = [ { 'n_estimators' : [ 3 , 10 , 30, 50 ], 'max_features' : [ 2 , 4 , 6 , 8, 10]}, 
              { 'bootstrap' : [ False ], 'n_estimators' : [ 3 , 10 , 15], 'max_features' : [ 2 , 3 , 4 , 5]}, ]
forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)
grid_search.fit(housing_prepared, housing_labels)
```
